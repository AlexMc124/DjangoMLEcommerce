```python
import numpy as np, pandas as pd, warnings
warnings.filterwarnings('ignore')
```

```python
reviews = pd.read_csv('data/ratings.csv')
```

```python
reviews.head()
```

```python
items = pd.read_csv('data/items.csv')
items.head()
```

```python
reviews = pd.merge(reviews,items,on='itemID')
reviews
```

```python
reviews.rename(columns={"userId": "UserID", "itemID": "ItemID"}, inplace=True)
```

```python
reviews.describe()
```

Mean rating is 3.501

rename the column names

```python
ratings = pd.DataFrame(reviews.groupby('itemTitle')['rating'].mean())
ratings.head()
```

```python
ratings
```

```python
ratings['number_of_ratings'] = reviews.groupby('itemTitle')['rating'].count()
ratings
```

```python
ratings['number_of_ratings'].max()
```

many ratings have 0 ratings

User Item Interaction Matrix 

```python
UII_Matrix = reviews.pivot_table(index='UserID', columns='itemTitle', values='rating')
UII_Matrix
```

## Making Reccomendation - ITEMS using Item 285 as example

```python
#Fetch Ratings for Item 
Test_item_rating = UII_Matrix['Item 258']
Test_item_rating
```

```python
# Finding the correlation with different Items 
similar_to_item = UII_Matrix.corrwith(Test_item_rating)
similar_to_item
```

```python
#rank ascending we would expect the result to be item 258
x = similar_to_item.sort_values(ascending=False)
```
